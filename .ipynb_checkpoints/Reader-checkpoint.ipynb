```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as ccrs
```

```python
def get_lonlat():
    
    """reads in 25 km polar sterographic longitude and latitude grids from binary
    and returns a dictionary of the two grids."""
    
    lat_path, lon_path = 'psn25lats_v3.dat', 'psn25lons_v3.dat'
    
    with open(lat_path,'rb') as fin:
    
        lats = [int.from_bytes(fin.read(4), 'little')/100_000 for i in range(136192)]
    
        lats_grid = np.reshape(lats, (448,304))
        
    with open(lon_path,'rb') as fin:
    
        lons = [int.from_bytes(fin.read(4), 'little',signed=True)/100_000 for i in range(136192)]
    
        lons_grid = np.reshape(lons, (448,304))
        
    return({'lons':lons_grid,
            'lats':lats_grid})
    

def read(year,month,day,directory='data'):
    
    """reads ice concentration binary file and extracts the header info and the values.
    Then calls the function that gets the lon/lat grids and rolls all four together 
    to return a dictionary. Currently reads Nasa Team data, but nt can be changed to 
    bt (and version num changed) to read bootstrap, etc."""

    month = str(int(month)).zfill(2)
    day = str(int(day)).zfill(2)

    path = f'{directory}/nt_{year}{month}{day}_n07_v1.1_n.bin'
    header_size=300
    
    with open(path,'rb') as fin:
        
        header = fin.read(300)

        data = fin.read(136192)

        ints = [int(x) for x in data]

        grid = np.reshape(ints, (448,304))

    lons, lats = get_lonlat()['lons'], get_lonlat()['lats']
    
    return({'head':header,
            'conc':grid,
            'lons':lons,
            'lats':lats})
    
    
def plot(year, month, day):
    
    """plots concentration field on a given day using cartopy"""
    
    data = read(year,month,day)
    
    fig = plt.figure(figsize=(8,8))
    ax = plt.axes(projection=ccrs.NorthPolarStereo())
    
    ax.set_extent([-180, 180, 90, 66], ccrs.PlateCarree())
    
    masked_data = np.ma.masked_where(np.isin(x,[0,251,253,255,254]), x)
    masked_data = np.ma.filled(x,np.nan)
    
    ax.add_feature(cartopy.feature.LAND, edgecolor='black',zorder=1)
    
    ax.coastlines(resolution='110m')
    
    plt.pcolormesh(data['lons'], data['lats'], masked_data, vmin = 0, vmax = 250,
                 transform=ccrs.PlateCarree(),zorder=0,cmap='plasma')
    
    plt.colorbar()
    
    plt.show()

plot(1978,11,11)
```
