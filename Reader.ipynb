```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as ccrs
import xarray as xr
```

```python
def hemisphere_code(hemisphere):
    if hemisphere.lower() == 'north': hem = 'n'; dims = (448,304)
    elif hemisphere.lower() == 'south': hem = 's'; dims = (332, 316)
    else:
        raise Exception('You must specify the hemisphere in the format "north"/"south"')
    return(hem, dims)

def get_lonlat(hemisphere):
    
    """reads in 25 km polar sterographic longitude and latitude grids from binary
    and returns a dictionary of the two grids."""
    
    hem, dims = hemisphere_code(hemisphere)
    
    lat_path, lon_path = f'ps{hem}25lats_v3.dat', f'ps{hem}25lons_v3.dat'
    
    with open(lat_path,'rb') as fin:
    
        lats = [int.from_bytes(fin.read(4), 'little',signed=True)/100_000 for i in range(dims[0]*dims[1])]
    
        lats_grid = np.reshape(lats, dims)
        
    with open(lon_path,'rb') as fin:
    
        lons = [int.from_bytes(fin.read(4), 'little',signed=True)/100_000 for i in range(dims[0]*dims[1])]
    
        lons_grid = np.reshape(lons, dims)
        
    return({'lons':lons_grid,
            'lats':lats_grid})
    

def read(year,month,day,hemisphere,directory='data'):
    
    """reads ice concentration binary file and extracts the header info and the values.
    Then calls the function that gets the lon/lat grids and rolls all four together 
    to return a dictionary. Currently reads Nasa Team data, but nt can be changed to 
    bt (and version num changed) to read bootstrap, etc."""
    
    hem, dims = hemisphere_code(hemisphere)

    month = str(int(month)).zfill(2)
    day = str(int(day)).zfill(2)

    path = f'{directory}/nt_{year}{month}{day}_n07_v1.1_{hem}.bin'
    header_size=300
    
    with open(path,'rb') as fin:
        
        header = fin.read(300).decode('UTF-8')
        
        data = fin.read(dims[0]*dims[1])

        ints = [int(x) for x in data]

        grid = np.reshape(ints, dims).astype(float)

    lons, lats = (get_lonlat(hemisphere)['lons'],
                  get_lonlat(hemisphere)['lats'])
    
    return({'head':header,
            'conc':grid,
            'lons':lons,
            'lats':lats})
    
    
def plot(year, month, day, hemisphere):
    
    """plots concentration field on a given day using cartopy"""
    
    data = read(year,month,day, hemisphere)
    
    fig = plt.figure(figsize=(8,8))
    
    if hemisphere.lower() == 'north':
        ax = plt.axes(projection=ccrs.NorthPolarStereo())
        ax.set_extent([-180, 180, 90, 66], ccrs.PlateCarree())
    elif hemisphere.lower() == 'south':        
        ax = plt.axes(projection=ccrs.SouthPolarStereo())
        ax.set_extent([-180, 180, -50, -90], ccrs.PlateCarree())        

    
    masked_data = np.ma.masked_where(np.isin(data['conc'],[0,251,253,255,254]), data['conc'])
    masked_data = np.ma.filled(masked_data, np.nan)/2.5
    
    ax.add_feature(cartopy.feature.LAND, edgecolor='black',zorder=1)
    
    ax.coastlines(resolution='110m')
        
    plt.pcolormesh(data['lons'], data['lats'], masked_data, vmin = 0, vmax = 100,
                 transform=ccrs.PlateCarree(),zorder=0,cmap='plasma')
    
    plt.colorbar()
    
    plt.show()

plot(1978,11,11,'north')
plot(1978,11,13,'south')
```
